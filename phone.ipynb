```python
import os
import json
import pandas as pd

# Path to insurance JSON data
path_3 = r"D:\phone_pe\data\data\aggregated\insurance\country\india\state"
Agg_insurance_state_list = os.listdir(path_3)

# Empty dictionary create
Agg_insurance = {
    'State': [], 'Year': [], 'Quarter': [],
    'insurance_type': [], 'insurance_count': [], 'insurance_amount': []
}

# Loop
for i in Agg_insurance_state_list:
    p_i = os.path.join(path_3, i)  # Example: tamil-nadu
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Example: 2018, 2019, 2020...
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Example: 1.json, 2.json, 3.json...
            with open(p_k, 'r') as Data:
                C = json.load(Data)

                # Safely check if "transactionData" exists
                if C.get("data") and "transactionData" in C["data"]:
                    for l in C["data"]["transactionData"]:
                        name = l["name"]  # Insurance type (ex: Insurance)
                        count = l["paymentInstruments"][0]["count"]
                        amount = l["paymentInstruments"][0]["amount"]

                        Agg_insurance["State"].append(i)
                        Agg_insurance["Year"].append(j)
                        Agg_insurance["Quarter"].append(int(k.strip('.json')))
                        Agg_insurance["insurance_type"].append(name)
                        Agg_insurance["insurance_count"].append(count)
                        Agg_insurance["insurance_amount"].append(amount)

# Convert into DataFrame
df_aggregated_insurance = pd.DataFrame(Agg_insurance)

pd.DataFrame(Agg_insurance)
```

```python
import os
import json
import pandas as pd

# Path to map transaction data
path_4 = r"D:\phone_pe\data\data\map\transaction\hover\country\india\state"
map_trans_state_list = os.listdir(path_4)

# Empty dictionary
map_trans = {
    'State': [], 'Year': [], 'Quarter': [],
    'District': [], 'Transaction_count': [], 'Transaction_amount': []
}

# Loop through states, years, quarters
for i in map_trans_state_list:
    p_i = os.path.join(path_4, i)  # Example: tamil-nadu
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Example: 2018, 2019, 2020...
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Example: 1.json, 2.json, 3.json...
            with open(p_k, 'r') as Data:
                D = json.load(Data)

            # Extract hoverDataList (districts)
            if D.get("data") and "hoverDataList" in D["data"]:
                for l in D["data"]["hoverDataList"]:
                    district = l["name"]
                    count = l["metric"][0]["count"]
                    amount = l["metric"][0]["amount"]

                    map_trans['State'].append(i)
                    map_trans['Year'].append(j)
                    map_trans['Quarter'].append(int(k.strip('.json')))
                    map_trans['District'].append(district)
                    map_trans['Transaction_count'].append(count)
                    map_trans['Transaction_amount'].append(amount)

# Convert into DataFrame
df_map_transaction = pd.DataFrame(map_trans)

pd.DataFrame(map_trans)
```

```python
import os
import json
import pandas as pd

# Path to map user hover data
path_5 = r"D:\phone_pe\data\data\map\user\hover\country\india\state"
map_user_state_list = os.listdir(path_5)

# Empty dictionary
map_user = {
    'State': [], 'Year': [], 'Quarter': [],
    'District': [], 'Registered_Users': [], 'App_Opens': []
}

# Loop through states, years, quarters
for i in map_user_state_list:
    p_i = os.path.join(path_5, i)  # Example: tamil-nadu
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Example: 2018, 2019, 2020...
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Example: 1.json, 2.json, 3.json...
            with open(p_k, 'r') as Data:
                E = json.load(Data)

            # Extract hoverData (district/state level users data)
            if E.get("data") and "hoverData" in E["data"]:
                for district, values in E["data"]["hoverData"].items():
                    registered = values.get("registeredUsers", 0)
                    app_opens = values.get("appOpens", 0)

                    map_user['State'].append(i)
                    map_user['Year'].append(j)
                    map_user['Quarter'].append(int(k.strip('.json')))
                    map_user['District'].append(district)
                    map_user['Registered_Users'].append(registered)
                    map_user['App_Opens'].append(app_opens)

# Convert into DataFrame
df_map_user = pd.DataFrame(map_user)

pd.DataFrame(map_user)
```

```python
import os
import json
import pandas as pd

# Path to top transaction data
path_top = r"D:\phone_pe\data\data\top\transaction\country\india\state"
top_trans_state_list = os.listdir(path_top)

# Empty dictionary
top_trans = {
    'State': [], 'Year': [], 'Quarter': [],
    'Entity': [], 'Entity_type': [], 'Transaction_count': [], 'Transaction_amount': []
}

# Loop through states, years, quarters
for i in top_trans_state_list:
    p_i = os.path.join(path_top, i)  # Example: tamil-nadu
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = os.path.join(p_i, j)  # Example: 2018, 2019, 2020...
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)  # Example: 1.json, 2.json, 3.json...
            with open(p_k, 'r') as Data:
                G = json.load(Data)

            if G.get("data"):
                # Loop over states, districts, pincodes
                for category in ["states", "districts", "pincodes"]:
                   if category in G["data"] and G["data"][category]:
                        for l in G["data"][category]:
                            entity = l["entityName"]
                            count = l["metric"]["count"]
                            amount = l["metric"]["amount"]

                            top_trans['State'].append(i)
                            top_trans['Year'].append(j)
                            top_trans['Quarter'].append(int(k.strip('.json')))
                            top_trans['Entity'].append(entity)
                            top_trans['Entity_type'].append(category)
                            top_trans['Transaction_count'].append(count)
                            top_trans['Transaction_amount'].append(amount)

# Convert into DataFrame
df_top_transaction = pd.DataFrame(top_trans)

pd.DataFrame(top_trans)
```

```python
import os
import json
import pandas as pd

# Path to top user data
path_top_user = r"D:\phone_pe\data\data\top\user\country\india\state"
top_user_state_list = os.listdir(path_top_user)

# Empty dictionary
top_user = {
    'State': [], 'Year': [], 'Quarter': [],
    'Entity': [], 'Entity_type': [], 'Registered_Users': []
}

# Loop through states, years, quarters
for i in top_user_state_list:
    p_i = os.path.join(path_top_user, i)
    years = os.listdir(p_i)

    for j in years:
        p_j = os.path.join(p_i, j)
        quarters = os.listdir(p_j)

        for k in quarters:
            p_k = os.path.join(p_j, k)
            with open(p_k, 'r') as f:
                H = json.load(f)

            if H.get("data"):
                for category in ["states", "districts", "pincodes"]:
                    if category in H["data"] and H["data"][category]:  # Check not None
                        for item in H["data"][category]:
                            name = item.get("name")
                            users = item.get("registeredUsers", 0)

                            top_user['State'].append(i)
                            top_user['Year'].append(j)
                            top_user['Quarter'].append(int(k.strip('.json')))
                            top_user['Entity'].append(name)
                            top_user['Entity_type'].append(category)
                            top_user['Registered_Users'].append(users)

# Convert to DataFrame
df_top_user = pd.DataFrame(top_user)

pd.DataFrame(top_user)
```

```python
# import os
# import json
# import pandas as pd

# Path to top insurance JSON data
path_top_insurance = r"D:\phone_pe\data\data\top\insurance\country\india\state"
top_insurance_state_list = os.listdir(path_top_insurance)

# Empty dictionary
top_insurance = {
    'State': [], 'Year': [], 'Quarter': [],
    'Entity': [], 'Entity_type': [], 'Insurance_Count': [], 'Insurance_Amount': []
}

# Loop through states, years, quarters
for state in top_insurance_state_list:
    p_state = os.path.join(path_top_insurance, state)
    years = os.listdir(p_state)

    for year in years:
        p_year = os.path.join(p_state, year)
        quarters = os.listdir(p_year)

        for quarter_file in quarters:
            p_quarter = os.path.join(p_year, quarter_file)
            with open(p_quarter, 'r') as f:
                I = json.load(f)

            if I.get("data"):
                for category in ["states", "districts", "pincodes"]:
                    if category in I["data"] and I["data"][category]:  # Check not None
                        for item in I["data"][category]:
                            name = item.get("entityName")
                            metric = item.get("metric", {})
                            count = metric.get("count", 0)
                            amount = metric.get("amount", 0)

                            top_insurance['State'].append(state)
                            top_insurance['Year'].append(year)
                            top_insurance['Quarter'].append(int(quarter_file.strip('.json')))
                            top_insurance['Entity'].append(name)
                            top_insurance['Entity_type'].append(category)
                            top_insurance['Insurance_Count'].append(count)
                            top_insurance['Insurance_Amount'].append(amount)

# Convert to DataFrame
df_top_insurance = pd.DataFrame(top_insurance)

pd.DataFrame(top_insurance)
```
